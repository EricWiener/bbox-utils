# Using bbox-utils to visualize bounding boxes within images

```python
from bbox_utils.image import Image
from bbox_utils import BoundingBox
```

### Create an object from a file
We can then display the image using a library of our choice (as long as you choose matplotlib or opencv).

```python
img = Image.load_from_file("./assets/buoys.png")
```

```python
img.display()
```

### Display bounding boxes on an image

```python
# Load the annotations
annotations = [
    { "class": 0, "xyxy": [288, 225, 294, 235] },
    { "class": 0, "xyxy": [338, 224, 342, 231] },
    { "class": 0, "xyxy": [360, 224, 368, 230] },
    { "class": 0, "xyxy": [401, 224, 408, 231] },
    { "class": 1, "xyxy": [248, 226, 256, 233] },
    { "class": 3, "xyxy": [237, 227, 244, 233] },
    { "class": 3, "xyxy": [293, 226, 298, 232] },
    { "class": 3, "xyxy": [325, 224, 330, 230] },
    { "class": 3, "xyxy": [386, 224, 394, 231] },
    { "class": 4, "xyxy": [276, 223, 288, 232] },
    { "class": 7, "xyxy": [372, 206, 382, 230] },
    { "class": 8, "xyxy": [1, 220, 133, 240] }
]

# Colors specified in BGR
color_map = {
    0: (208, 114, 244), # Pink
    1: (10, 163, 240), # Amber
    3: (0, 200, 227), # Yellow
    4: (23, 169, 96), # Green
    7: (226, 161, 27), # Cyan
    8: (255, 0, 170), # Violet
}
```

```python
bboxes = []
colors = []

# Create BoundingBox objects
for annotation in annotations:
    x1, y1 = annotation["xyxy"][:2]
    x2, y2 = annotation["xyxy"][2:]
    bboxes.append(BoundingBox.from_xyxy((x1, y1), (x2, y2)))
    colors.append(color_map[annotation["class"]])
```

You can either display a single bounding box:

```python
img.display_bbox(bboxes[0], colors=colors[0])
```

Or a list of bounding boxes

```python
img.display_bboxes(bboxes, colors=colors)
```
